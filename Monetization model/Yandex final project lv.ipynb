```python
''' –§—É–Ω–∫—Ü–∏—è plot_func –≥—Ä—É–ø–ø–∏—Ä—É–µ—Ç –¥–∞—Ç–∞—Ñ—Ä–µ–π–º (df) –ø–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–º—É –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ —Å—Ç–æ–ª–±—Ü—É(group_column),
        –ø—Ä–∏–º–µ–Ω–∏—è –∞–≥–≥—Ä–µ–≥–∏—Ä—É—é—â—É—é —Ñ—É–Ω–∫—Ü–∏—é (aggfunc). –ù–∞ –≤—Ö–æ–¥ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ–¥–∞–µ—Ç—Å—è –∑–∞–≥–æ–ª–æ–≤–æ–∫ –≥—Ä–∞—Ñ–∏–∫–∞ (figure_name),
        –∞ —Ç–∞–∫–∂–µ –Ω–∞–∑–≤–∞–Ω–∏—è –æ—Å–µ–π x –∏ —É —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ (xlabel_name, ylabel_name) '''

def plot_func(df, figure_name, xlabel_name,ylabel_name,
             agg_func, group_column):
    new_df = eval('df.groupby(group_column).{}()'.format(agg_func))
    new_col_name = ''
    for x in new_df.columns:
        new_col_name = x
    new_df = new_df.sort_values(by=new_col_name)
    colors = ['aquamarine', 'violet', 'wheat', 'cornflowerblue']
    xticks = [re.sub('_+',' ',x).capitalize() for x in new_df.index]
    plt.figure(figsize=(8,8))
    plt.bar(new_df[new_col_name].index, new_df[new_col_name], color=colors,ec='black'
           ,alpha=0.5)
    plt.title(figure_name)
    plt.xticks(new_df.index,xticks)
    plt.grid(color='black', alpha = 0.1, axis = 'y', linestyle = ':')
    plt.tick_params(direction='inout')
    plt.xlabel(xlabel_name)
    plt.ylabel(ylabel_name)
    plt.yticks(np.arange (0, round(new_df[new_col_name].max()/500 + 1.1)*500, 500))
    sns.despine()
    plt.show()
    

''' –§—É–Ω–∫—Ü–∏—è hist_func –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ —Å—Ç–æ–ª–±–µ—Ü –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ (series), –ø–æ –∫–æ—Ç–æ—Ä–æ–º—É –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ.
        –ù–∞ –≤—Ö–æ–¥ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ–¥–∞–µ—Ç—Å—è –∑–∞–≥–æ–ª–æ–≤–æ–∫ –≥—Ä–∞—Ñ–∏–∫–∞ (figure_name),–∞ —Ç–∞–∫–∂–µ –Ω–∞–∑–≤–∞–Ω–∏—è –æ—Å–µ–π x –∏ —É 
        —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ (xlabel_name, ylabel_name) '''

def hist_func(series, figure_name, xlabel_name,ylabel_name):
    plt.figure(figsize=(8,8))
    plt.hist(series, histtype='bar', ec='black', bins=50, color='aquamarine',alpha=0.5)
    plt.grid(color='black', alpha = 0.1, axis = 'y', linestyle = ':')
    sns.despine()
    plt.ylabel(ylabel_name)
    plt.xlabel(xlabel_name)
    plt.title(figure_name)
    plt.tick_params(direction='inout')
    plt.show()
    
    
''' –§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º, —Å–æ–¥–µ—Ä–∂–∞—â–∏–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –¥–µ–π—Å—Ç–≤–∏—è—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç
        —Ç–∞–±–ª–∏—Ü—É —Å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–º–∏ –ø—Ä–æ—Ñ–∏–ª—è–º–∏ '''

def get_profiles(sessions):
    profiles = (
        sessions.sort_values(by=['user_id', 'event_datetime'])
        .groupby('user_id')
        .agg({'date': 'first', 'source': 'first', 'category':'first'})
        .rename(columns={'date': 'first_action'})
        .reset_index()  # –≤–æ–∑–≤—Ä–∞—â–∞–µ–º user_id –∏–∑ –∏–Ω–¥–µ–∫—Å–∞
    )
    profiles['month'] = profiles['first_action'].astype('datetime64[M]')

    return profiles


''' –§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –¥–≤–∞ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞. –° —Ç–∞–±–ª–∏—Ü–µ–π –ø—Ä–æ—Ñ–∏–ª–µ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ —Å —Ç–∞–±–ª–∏—Ü–µ–π, —Å–æ–¥–µ—Ä–∂–∞—â–µ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é 
–æ –ø—Ä–∏–±—ã–ª—å–Ω–æ–º –¥–µ–π—Å—Ç–≤–∏–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è. –ù–∞ –≤—ã—Ö–æ–¥–µ –≤—ã–¥–∞—é—Ç—Å—è —Ç–∞–±–ª–∏—Ü—ã LTV –∏ ROI'''
def get_ltv(
    profiles,
    purchases,
    dimensions=[]
):

    result_raw = profiles
    # –¥–æ–±–∞–≤–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –æ –ø–æ–∫—É–ø–∫–∞—Ö –≤ –ø—Ä–æ—Ñ–∏–ª–∏
    result_raw = result_raw.merge(
        purchases[['user_id', 'date', 'click_cost','acquisition_cost']], on='user_id', how='left'
    )
    # —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –ª–∞–π—Ñ—Ç–∞–π–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –¥–ª—è –∫–∞–∂–¥–æ–π –ø–æ–∫—É–ø–∫–∏
    result_raw['lifetime'] = (
        result_raw['date'] - result_raw['first_action']
    ).dt.days
    # –≥—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ cohort, –µ—Å–ª–∏ –≤ dimensions –Ω–∏—á–µ–≥–æ –Ω–µ—Ç
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort']

    # —Ñ—É–Ω–∫—Ü–∏—è –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∏ –ø–æ –∂–µ–ª–∞–µ–º—ã–º –ø—Ä–∏–∑–Ω–∞–∫–∞–º
    def group_by_dimensions(df, dims):
        # —Å—Ç—Ä–æ–∏–º ¬´—Ç—Ä–µ—É–≥–æ–ª—å–Ω—É—é¬ª —Ç–∞–±–ª–∏—Ü—É –≤—ã—Ä—É—á–∫–∏
        result = df.pivot_table(
            index=dims, columns='lifetime', values='click_cost', aggfunc='sum'
        )
        # –Ω–∞—Ö–æ–¥–∏–º —Å—É–º–º—É –≤—ã—Ä—É—á–∫–∏ —Å –Ω–∞–∫–æ–ø–ª–µ–Ω–∏–µ–º
        result = result.fillna(0).cumsum(axis=1)
        # –≤—ã—á–∏—Å–ª—è–µ–º —Ä–∞–∑–º–µ—Ä—ã –∫–æ–≥–æ—Ä—Ç
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        # –æ–±—ä–µ–¥–∏–Ω—è–µ–º —Ä–∞–∑–º–µ—Ä—ã –∫–æ–≥–æ—Ä—Ç –∏ —Ç–∞–±–ª–∏—Ü—É –≤—ã—Ä—É—á–∫–∏
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # —Å—á–∏—Ç–∞–µ–º LTV: –¥–µ–ª–∏–º –∫–∞–∂–¥—É—é ¬´—è—á–µ–π–∫—É¬ª –≤ —Å—Ç—Ä–æ–∫–µ –Ω–∞ —Ä–∞–∑–º–µ—Ä –∫–æ–≥–æ—Ä—Ç—ã
        result = result.div(result['cohort_size'], axis=0)
        result['cohort_size'] = cohort_sizes

        # —Å–æ—Ö—Ä–∞–Ω—è–µ–º –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –¥–∞–Ω–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –∑–Ω–∞—á–µ–Ω–∏—è CAC, 
        # –¥–æ–±–∞–≤–∏–≤ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –∏–∑ dimensions
        cac = df[['user_id', 'acquisition_cost'] + dims].drop_duplicates()

        # —Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–∏–π CAC –ø–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º –∏–∑ dimensions
        cac = (
            cac.groupby(dims)
            .agg({'acquisition_cost': 'mean'})
            .rename(columns={'acquisition_cost': 'cac'})
        )

        # —Å—á–∏—Ç–∞–µ–º ROI: –¥–µ–ª–∏–º LTV –Ω–∞ CAC
        roi = result.div(cac['cac'], axis=0)


        # –≤–æ—Å—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º —Ä–∞–∑–º–µ—Ä—ã –∫–æ–≥–æ—Ä—Ç –≤ —Ç–∞–±–ª–∏—Ü–µ ROI
        roi['cohort_size'] = cohort_sizes

        # –¥–æ–±–∞–≤–ª—è–µ–º CAC –≤ —Ç–∞–±–ª–∏—Ü—É ROI
        roi['cac'] = cac['cac']

        # –≤–æ–∑–≤—Ä–∞—â–∞–µ–º —Ç–∞–±–ª–∏—Ü—ã LTV –∏ ROI
        return result, roi

    # –ø–æ–ª—É—á–∞–µ–º —Ç–∞–±–ª–∏—Ü—ã LTV –∏ ROI
    result_grouped, roi_grouped = group_by_dimensions(
        result_raw, dimensions
    )

    # –¥–ª—è —Ç–∞–±–ª–∏—Ü –¥–∏–Ω–∞–º–∏–∫–∏ —É–±–∏—Ä–∞–µ–º 'cohort' –∏–∑ dimensions
    if 'cohort' in dimensions:
        dimensions = []

    # –ø–æ–ª—É—á–∞–µ–º —Ç–∞–±–ª–∏—Ü—ã –¥–∏–Ω–∞–º–∏–∫–∏ LTV –∏ ROI
    result_in_time, roi_in_time = group_by_dimensions(
        result_raw, dimensions + ['first_action']
    )

    return (
        result_raw,  # —Å—ã—Ä—ã–µ –¥–∞–Ω–Ω—ã–µ
        result_grouped,  # —Ç–∞–±–ª–∏—Ü–∞ LTV
        result_in_time,  # —Ç–∞–±–ª–∏—Ü–∞ –¥–∏–Ω–∞–º–∏–∫–∏ LTV
        roi_grouped,  # —Ç–∞–±–ª–∏—Ü–∞ ROI
        roi_in_time,  # —Ç–∞–±–ª–∏—Ü–∞ –¥–∏–Ω–∞–º–∏–∫–∏ ROI
    )


''' –§—É–Ω–∫—Ü–∏—è vis_ltv —Å—Ç—Ä–æ–∏—Ç –ª–∏–Ω–µ–π–Ω—ã–µ –≥—Ä–∞—Ñ–∏–∫–∏ ltv –∫–æ–≥–æ—Ä—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π'''

def vis_ltv(df, title):
    fig,ax = plt.subplots()
    df.plot(figsize=(10, 5), xticks=list(report.columns.values), ax=ax, linewidth=2)
    plt.title(title)
    plt.ylabel('LTV, $')
    plt.xlabel('–õ–∞–π—Ñ—Ç–∞–π–º')
    colormap = plt.cm.gist_rainbow
    colors = [colormap(i) for i in np.linspace(0, 1,len(ax.lines))]
    for i,j in enumerate(ax.lines):
        j.set_color(colors[i])
    plt.grid(color='black', linestyle='dotted', alpha = 0.5)
    plt.tick_params(direction='inout')
    sns.despine()
    plt.legend()
    plt.show()
    
    
''' –§—É–Ω–∫—Ü–∏—è vis_roi —Å—Ç—Ä–æ–∏—Ç –ª–∏–Ω–µ–π–Ω—ã–µ –≥—Ä–∞—Ñ–∏–∫–∏ ROMI '''  

def vis_roi(df, title):
    fig,ax = plt.subplots()
    df.T.plot(figsize=(10, 5), xticks=list(df.columns.values), ax=ax, linewidth=2)
    plt.title(title)
    plt.ylabel('ROMI')
    plt.xlabel('–õ–∞–π—Ñ—Ç–∞–π–º')
    colormap = plt.cm.gist_rainbow
    colors = [colormap(i) for i in np.linspace(0, 1,len(ax.lines))]
    for i,j in enumerate(ax.lines):
        j.set_color(colors[i])
    plt.grid(color='black', linestyle='dotted', alpha = 0.5)
    plt.tick_params(direction='inout')
    sns.despine()
    plt.axhline(y=1, color='black', linestyle='--', label='–£—Ä–æ–≤–µ–Ω—å –æ–∫—É–ø–∞–µ–º–æ—Å—Ç–∏')
    plt.legend()
    plt.show()
```

### –ì–∏–ø–æ—Ç–µ–∑–∞ –æ —Ä–∞–∑–ª–∏—á–∏–∏ –¥–æ–ª–∏ –ø—Ä–∏–±—ã–ª–∏, –ø—Ä–∏–Ω–æ—Å–∏–º–æ–π –∏–≥—Ä–æ–∫–∞–º–∏ –∏–∑ —Ä–∞–∑–Ω—ã—Ö –∫–∞–Ω–∞–ª–æ–≤

–°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –≥–∏–ø–æ—Ç–µ–∑—ã
–ù—É–ª–µ–≤–∞ –≥–∏–ø–æ—Ç–µ–∑–∞—è (H0): –î–æ–ª—è –æ—Ç –ø—Ä–∏–±—ã–ª–∏ –∏–≥—Ä–æ–∫–æ–≤ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–Ω—ã—Ö –∏–∑ –∫–∞–Ω–∞–ª–∞ Yandex direct —Ä–∞–≤–Ω–∞ –¥–æ–ª–µ –æ—Ç –ø—Ä–∏–±—ã–ª–∏ –∏–≥—Ä–æ–∫–æ–≤ –∏–∑ –∫–∞–Ω–∞–ª–∞ Youtube channel reklama –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –∏–≥—Ä–æ–∫–æ–≤, –∑–∞–≤–µ—Ä—à–∏–≤—à–∏—Ö —É—Ä–æ–≤–µ–Ω—å –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ–º –ø—Ä–æ–µ–∫—Ç–∞

–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞(H1): –ú–µ–∂–¥—É –¥–æ–ª—è–º–∏ –æ—Ç –ø—Ä–∏–±—ã–ª–∏ –∏–≥—Ä–æ–∫–æ–≤ –∏–∑ –∫–∞–Ω–∞–ª–æ–≤ Yandex direct –∏ Youtube channel reklama –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –∏–≥—Ä–æ–∫–æ–≤, –∑–∞–≤–µ—Ä—à–∏–≤—à–∏—Ö –ø–µ—Ä–≤—ã–π —É—Ä–æ–≤–µ–Ω—å –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ–º –ø—Ä–æ–µ–∫—Ç–∞, —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ–µ —Ä–∞–∑–ª–∏—á–∏–µ

–£—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ = 0.05

```python
alpha = .05 # –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
successes1 = ga_with_category.query(
    ('source == "yandex_direct" and category == "finished by project"'))['click_cost'].sum()
successes2 = ga_with_category.query(
    ('source == "youtube_channel_reklama" and category == "finished by project"'))['click_cost'].sum()
trials = ga_with_category.query('category == "finished by project"')['click_cost'].sum()

#–¥–æ–ª—è —É—Å–ø–µ—Ö–æ–≤ –≤ –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ
p = (successes1 + successes2) / trials
     
#—Ä–∞–∑–Ω–∏—Ü–∞ –¥–æ–ª–µ–π –≤ –¥–∞—Ç–∞—Å–µ—Ç–∞—Ö
difference = successes1/trials - successes2/trials
     
z_value = difference/mt.sqrt(p*(1-p)*(1/trials + 1/trials))

distr = st.norm(0,1)
     
p_value = (1 - distr.cdf(abs(z_value))) * 2

if p_value < alpha: 
    print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', p_value), print("–ú–µ–∂–¥—É –≤—ã–±–æ—Ä–∫–∞–º–∏ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–∞—è —Ä–∞–∑–Ω–∏—Ü–∞")
else: 
    print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', p_value), print("–ù–µ—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ–π —Ä–∞–∑–Ω–∏—Ü—ã –º–µ–∂–¥—É –≤—ã–±–æ—Ä–∫–∞–º–∏")
```
